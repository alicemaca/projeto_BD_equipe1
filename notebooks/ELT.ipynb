```python
import pandas as pd
from sqlalchemy import create_engine

usuario_db = 'postgres'
senha_db = ''
host_db = 'localhost'
porta_db = '5432'
nome_banco_db = 'elt_enem'

DATABASE_URL = f'postgresql://{usuario_db}:{senha_db}@{host_db}:{porta_db}/{nome_banco_db}'
engine = create_engine(DATABASE_URL)

print("Conexão criada.")

arquivos_brutos = {
    'MICRODADOS_ENEM_2021': 'dados/MICRODADOS_ENEM_2021_CORTADO.csv',
    'MICRODADOS_ENEM_2022': 'dados/MICRODADOS_ENEM_2022_CORTADO.csv',
    'MICRODADOS_ENEM_2023': 'dados/MICRODADOS_ENEM_2023_CORTADO.csv'
}
```

```python
print("Iniciando carga dos dados BRUTOS...")

for nome_tabela, caminho_arquivo in arquivos_brutos.items():
    try:
        print(f"Lendo o arquivo: {caminho_arquivo}")
        df_raw = pd.read_csv(
            caminho_arquivo, 
            encoding='utf-8',
            on_bad_lines='skip' 
        )
        
        print(f"Carregando {len(df_raw)} linhas brutos na tabela: {nome_tabela}...")
        
        df_raw.to_sql(
            nome_tabela,
            con=engine,
            if_exists='replace',
            index=False
        )
        print(f"Tabela {nome_tabela} carregada com sucesso.")
        
    except Exception as e:
        print(f"Erro ao processar {caminho_arquivo}: {e}")

print("\nProcesso de EL (Extração e Carga) concluído.")
```
